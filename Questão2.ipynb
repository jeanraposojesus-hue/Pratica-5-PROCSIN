### Exercício 2: Resumo sobre Transformadas de Wavelets

A **Transformada de Wavelet (WT)** é uma ferramenta matemática que decompõe um sinal em diferentes componentes de frequência com resoluções variadas. Diferente da Transformada de Fourier, que oferece resolução perfeita em frequência mas nenhuma no tempo, a WT utiliza janelas de tamanhos variáveis (wavelets-mãe escalonadas e transladadas).

**Principais características:**
- **Multirresolução:** Janelas estreitas para altas frequências (melhor resolução temporal) e janelas largas para baixas frequências (melhor resolução de frequência).
- **Localização:** Capaz de identificar transientes e descontinuidades.

**Aplicações:**
1. **Compressão de Imagem e Áudio:** Utilizada no padrão JPEG2000.
2. **Denoising (Remoção de Ruído):** Separa o ruído (geralmente em altas frequências/detalhes) do sinal principal.
3. **Engenharia Biomédica:** Análise de sinais de ECG e EEG para detecção de anomalias.
4. **Geofísica:** Processamento de dados sísmicos.
