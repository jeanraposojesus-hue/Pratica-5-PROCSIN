```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
from scipy import signal

def plot_spectrum(ax, x, fs, title, xlim=None):
    fft_x = np.fft.fft(x)
    freqs = np.fft.fftfreq(len(x), 1/fs)
    ax.plot(freqs[:len(x)//2], np.abs(fft_x[:len(x)//2]))
    ax.set_title(title)
    ax.set_xlabel('Frequência (Hz)')
    if xlim: ax.set_xlim(xlim)

def plot_wavelet_decomposition(x, coeffs, title):
    n = len(coeffs)
    fig, axes = plt.subplots(n + 1, 1, figsize=(10, 2*n))
    axes[0].plot(x)
    axes[0].set_title(title)
    for i, c in enumerate(coeffs):
        axes[i+1].plot(c)
        label = 'A5' if i == 0 else f'D{n-i}'
        axes[i+1].set_ylabel(label)
    plt.tight_layout()
    plt.show()

# 1. Geração do Sinal Não Estacionário
fs = 256
N = 2048
t = np.arange(N) / fs
x = np.zeros_like(t)

x[t < 2] = np.sin(2 * np.pi * 5 * t[t < 2])
x[(t >= 2) & (t < 4)] = signal.chirp(t[(t >= 2) & (t < 4)], f0=8, f1=28, t1=4, method='linear')
x[(t >= 4) & (t < 6)] = 0.7 * np.sin(2 * np.pi * 18 * t[(t >= 4) & (t < 6)]) * (1 + 0.6 * np.sin(2 * np.pi * 1.5 * t[(t >= 4) & (t < 6)]))
x[(t >= 6) & (t < 7)] = np.sin(2 * np.pi * 40 * t[(t >= 6) & (t < 7)])
x[int(6.4 * fs)] += 3.5  # Transiente
x[t >= 7] = 0.6 * np.sin(2 * np.pi * 10 * t[t >= 7]) + 0.4 * np.sin(2 * np.pi * 22 * t[t >= 7])

# 2. Decomposição Wavelet
wavelet = 'bior4.4'
level = 5
coeffs = pywt.wavedec(x, wavelet, level=level, mode='periodization')

# 3. Plotagem e Energia
plot_wavelet_decomposition(x, coeffs, f'Decomposição {wavelet} em {level} níveis')

print(f"{'Subfaixa':<10} | {'Energia':<15}")
print("-" * 30)
for i, c in enumerate(coeffs):
    label = 'A5' if i == 0 else f'D{level - i + 1}'
    energy = np.sum(np.square(c))
    print(f"{label:<10} | {energy:<15.4f}")
```
